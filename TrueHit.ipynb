```
import pandas as pd
import numpy as np

url = 'https://raw.githubusercontent.com/maniac73/Baseball/master/Truehit.csv'
df = pd.read_csv(url)
```

```
# This code find the Best TrueHit hitters in 2019
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
df.sort_values('TrueHit', ascending=False, inplace=True)
result = df[(df['yearID'] == 2019) & (df['G'] > 120)]
result.head(20)
```
